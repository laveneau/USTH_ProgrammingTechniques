## Exercise 1
Modify the following Python code by adding typing for parameters and return values.

```python
%%writefile .dummy/exercise1.py
import numpy as np

def power(x, y):
    result = 1.0
    while True:
        t = y % 2
        y = int(np.floor(y / 2))
        if t == 1:
            result = result * x
        if y == 0:
            break
        x = x * x
    return result

print(f"power(2,10) = {power(2, 10)}")
print(f"power(2.0,10) = {power(2.0, 10)}")
```

## Exercise 3
In this exercise, you will play with a linked list implementation taken from [https://realpython.com/linked-lists-python/](https://realpython.com/linked-lists-python/). 

The objective is obviously to add typing onto each parameters and returned value... 

```python
%%writefile .dummy/exercise3.py


class LinkedList:
    class __Node:
        def __init__(self, data, next = None):
            self.__data = data
            self.__next = next

        @property 
        def data(self):
            return self.__data
        
        @property
        def next(self):
            return self.__next
        
        @next.setter
        def next(self, _next):
            self.__next = _next

        def __repr__(self):
            return repr(self.data)

    def __init__(self, nodes = None):
        self.__head = None
        if nodes is not None:
            node = self.__Node(data=nodes.pop(0))
            self.__head = node
            for elem in nodes:
                node.next = self.__Node(elem)
                node = node.next

    def __repr__(self):
        node = self.__head
        nodes = []
        while node is not None:
            nodes.append(repr(node))
            node = node.next
        nodes.append("None")
        return " -> ".join(nodes)

    def __iter__(self):
        node = self.__head
        while node is not None:
            yield node.data
            node = node.next

    def is_empty(self):
        return self.__head is None
    
    def add_first(self, data):
        self.__head = self.__Node(data, self.__head)
        return self

    def add_last(self, data):
        if self.is_empty():
            return self.add_first(data)
            
        current_node = self.__head
        while current_node.next is not None:
            current_node = current_node.next
        current_node.next = self.__Node(data)
        
        return self
    
    def add_before(self, target_node_data, new_node_data):
        if self.is_empty():
            raise Exception("List is empty")

        if self.__head.data == target_node_data:
            return self.add_first(new_node_data)

        prev_node = self.__head
        node = prev_node.next
        while node is not None:
            if node.data == target_node_data:
                prev_node.next = self.__Node(new_node_data, node)
                return self
            prev_node = node
            node = node.next

        raise Exception(f"Node with data {target_node_data} not found")

    def add_after(self, target_node_data, new_node_data):
        if self.is_empty():
            raise Exception("List is empty")

        node = self.__head
        while node is not None:
            if node.data == target_node_data:
                node.next = self.__Node(new_node_data, node.next)
                return self
            node = node.next

        raise Exception(f"Node with data {target_node_data} not found")

    def remove_node(self, target_node_data):
        if self.is_empty():
            raise Exception("List is empty")

        if self.__head.data == target_node_data:
            self.__head = self.__head.next
            return self

        previous_node = self.__head
        node = previous_node.next
        while node is not None:
            if node.data == target_node_data:
                previous_node.next = node.next
                return self
            previous_node = node
            node = node.next

        raise Exception(f"Node with data {target_node_data} not found")


if __name__ == "__main__":
    def build_list():
        lklist = LinkedList(["b", "d", "f"])
        lklist.add_first("a")
        lklist.add_last("g")
        lklist.add_before("d", "c")
        lklist.add_after("d", "e")
        return lklist
    
    def remove_one_over_two(lklist):
        lklist.remove_node("a")
        lklist.remove_node("c")
        lklist.remove_node("e")
        lklist.remove_node("g")
        
    lklist = build_list()
    print(lklist)
    remove_one_over_two(lklist)
    print(lklist)
```
